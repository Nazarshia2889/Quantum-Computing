### Exercise #1

**Problem**: Print the product of the numbers in the list `numbers`.

````
numbers = [1, 5, 7, 3]
````

##### **Solution**

````
product = numbers[0]
for i in range(1, len(numbers)):
  product = product * numbers[i]
print(product)
````
